```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
import warnings
warnings.filterwarnings('ignore')

print("✓ All libraries ready!")
```

```python
sports = pd.DataFrame({
    'Sport':        ['Football', 'Squash', 'Padel', 'Esports', 'Pickleball', 'Rock Climbing'],
    'Start Year':   [1950,        1990,     2018,    2017,       2022,          2019],
    'Players 2024': [500000,      15000,    30000,   2500000,    3000,          5000],
    'Growth %':     [3,           5,        45,      30,         60,            40],
})

esports = pd.DataFrame({
    'Year':    [2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024],
    'Revenue': [50,   120,  200,  280,  450,  700,  1200, 2100],
})

padel = pd.DataFrame({
    'Year':    [2018, 2019, 2020, 2021, 2022, 2023, 2024],
    'Courts':  [20,   80,   200,  400,  700,  1000, 1200],
    'Players': [500,  2000, 5000, 10000, 18000, 25000, 30000],
})

print("✓ Data loaded!")
print(sports)
```

```python
plt.figure(figsize=(10, 6))

colors = ['#2563EB' if g >= 30 else '#9CA3AF' for g in sports['Growth %']]
bars = plt.bar(sports['Sport'], sports['Growth %'], color=colors, width=0.6)

for bar, val in zip(bars, sports['Growth %']):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.8,
             f'{val}%', ha='center', fontsize=10, fontweight='bold')

plt.axhline(30, color='red', linewidth=1.2, linestyle='--', label='30% threshold')
plt.title('Annual Growth Rate by Sport — Saudi Arabia', fontsize=14, fontweight='bold')
plt.ylabel('Growth Rate (%)')
plt.ylim(0, 75)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
model = LinearRegression()
model.fit(esports[['Year']], esports['Revenue'])
future_years = np.array(range(2025, 2031)).reshape(-1, 1)
forecast = model.predict(future_years)

plt.figure(figsize=(10, 6))
plt.fill_between(esports['Year'], esports['Revenue'], alpha=0.15, color='#7C3AED')
plt.plot(esports['Year'], esports['Revenue'], 'o-', color='#7C3AED', linewidth=2.5, markersize=7, label='Actual')
plt.plot(range(2025, 2031), forecast, 's--', color='#DC2626', linewidth=2, markersize=6, label='Forecast')
plt.title('Esports Revenue Growth & 2030 Forecast — Saudi Arabia', fontsize=14, fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Revenue (Million USD)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(10, 6))

ax2 = ax1.twinx()
ax1.bar(padel['Year'], padel['Courts'], color='#0D9488', alpha=0.7, width=0.4, label='Courts')
ax2.plot(padel['Year'], padel['Players'], 'o-', color='#D97706', linewidth=2.5, markersize=7, label='Players')

ax1.set_title('Padel Growth — Courts & Players — Saudi Arabia', fontsize=14, fontweight='bold')
ax1.set_xlabel('Year')
ax1.set_ylabel('Number of Courts', color='#0D9488')
ax2.set_ylabel('Registered Players', color='#D97706')
ax1.tick_params(axis='y', labelcolor='#0D9488')
ax2.tick_params(axis='y', labelcolor='#D97706')

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

plt.tight_layout()
plt.show()
```

```python
forecast_data = pd.DataFrame({
    'Sport':       ['Padel',  'Esports', 'Pickleball', 'Rock Climbing'],
    'Players2030': [150000,   40000000,  80000,        25000],
    'Growth %':    [45,       30,        60,           40],
    'Potential':   [8,        10,        7,            5],
})

colors = ['#0D9488', '#7C3AED', '#DC2626', '#D97706']

plt.figure(figsize=(10, 6))
for i, row in forecast_data.iterrows():
    plt.scatter(row['Growth %'], row['Players2030'],
                s=row['Potential'] * 150,
                color=colors[i], alpha=0.85, zorder=5)
    plt.annotate(row['Sport'],
                 (row['Growth %'], row['Players2030']),
                 textcoords='offset points', xytext=(10, 5),
                 fontsize=11, fontweight='bold')

plt.title('2030 Outlook — Emerging Sports in Saudi Arabia', fontsize=14, fontweight='bold')
plt.xlabel('Annual Growth Rate (%)')
plt.ylabel('Projected Players (2030)')
plt.gca().yaxis.set_major_formatter(
    plt.FuncFormatter(lambda x, _: f'{x/1e6:.0f}M' if x >= 1e6 else f'{x/1e3:.0f}K'))
plt.tight_layout()
plt.show()
```
